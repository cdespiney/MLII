## Part 1 : Data Distribution and the law of large numbers
### 1)

Where **x** represents the number of occurence of the letter 'A' in a book and is a random integer between 1 and 200
and **y** represents the height of the book in centimeter and is a 2 decimal float from 10 to 30.

Both **x** and **y** are discrete random variable.

```python
min_x = 0
max_x = 200
min_y = 10
max_y = 30
decimal_y = 2

# Each value of x and y have the same probability: 1 / nb of different values
prob_x = 1/(max_x - min_y)
prob_y = 1/((max_y - min_y) * np.power(10, decimal_y))

# expected value of x
expected_x = 0
for i in range(min_x, max_x):
    expected_x += i * prob_x

# expected value of x
expected_y = 0
for i in range(min_y, max_y * np.power(10, decimal_y)):
    expected_y += (i/100) * prob_y

print("The expected value of x is", expected_x)
print("The expected value of y is", expected_y)
```

### 2)

```python
n = 1000
x = np.random.randint(max_x, size=n)
y = np.round((max_y - min_y) * np.random.rand(n) + min_y, decimals=decimal_y) #rand from 10 to 30
#print(x)
#print(y)
fig, ax = plt.subplots()
ax.set_xlabel("number of occurence of the letter 'A'")
ax.set_ylabel("height of the book (in cm)")
ax.scatter(x, y)
```

#### 3)

```python
# Euclidian average for each samples of n
avgs_x = []
avgs_y = []
for i in range(1, n+1):
    avg_x = x[:i].sum()/i
    avg_y = y[:i].sum()/i
    avgs_x.append(avg_x)
    avgs_y.append(avg_y)

def euclidian_distance(x1, x2, y1, y2):
    return np.sqrt(np.square(x2-x1)+np.square(y2-y1))

# Euclidian distances
distances = []
for i in range(n):
    distances.append(euclidian_distance(avg_x, avgs_x[i], avg_y, avgs_y[i]))

# plot
fig, ax = plt.subplots()
ax.plot(range(1, n+1), distances)
```
