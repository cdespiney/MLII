```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

# Machine Learning: Unsupervised Learning, Agents
## Meteorological data : dimensionality reduction and visualization
### Method 1: Principal component analysis (PCA)

```python
data = np.load('./data/data.npy')
labels = np.load('./data/labels.npy')

# Colors list
colors = list(map(lambda label: (1, 0, 0) if label else (0, 0, 1), labels))

# 2D PCA dimension reduction
pca = PCA(n_components=2)
data_pca = pca.fit(data).transform(data)

plt.figure()
plt.scatter(data_pca[:, 0], data_pca[:, 1], color=colors)

# 3D PCA dimension reduction
pca = PCA(n_components=3)
data_pca = pca.fit(data).transform(data)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data_pca[:, 0], data_pca[:, 2], data_pca[:, 1], color=colors)
```

The prediction will be much more accurate with a dimension of 3

### Method 2: Multidimensional scaling (MDS)

```python
# 2D MDS dimension reduction
mds = MDS(n_components=2)
data_mds = mds.fit_transform(data)

plt.figure()
plt.scatter(data_mds[:, 0], data_mds[:, 1], color=colors)

# 3D MDS dimension reduction
mds = MDS(n_components=3)
data_mds = mds.fit_transform(data)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data_mds[:, 2], data_mds[:, 1], data_mds[:, 0], color=colors)
```

This method will have less accurate predictions regardless of the dimension.
